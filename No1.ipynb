```python
import pandas as pd
import os
```

```python
# Path folder tempat file CSV berada (misal: './data/')
folder_path = 'dataset/stock data'   
class2 = os.listdir(folder_path)


csv_files = []
stock_name = pd.read_csv('dataset/stock_code.csv')


for bla2 in class2:
    path2 = folder_path + "/" + bla2
    csv_files.append(path2)
    
```

```python
# List untuk menyimpan rata-rata volume
average_for_volumes = []
```

```python
print(csv_files)
```

```python
for x,file in enumerate(csv_files):

    # Membaca data dari setiap CSV
    df = pd.read_csv(file)

    df = df.drop(0)
    df = df.astype('float32')

    # Menghitung rata-rata volume
    yearly_volume_avg = df['Volume'].mean()

    file_name = file.split('/')[-1].split('.')[0]
    print(f'{file_name}: {yearly_volume_avg}')

```

```python
# Mengonversi list hasil ke DataFrame
result_df = pd.DataFrame(average_for_volumes)

# Pastikan kolom 'Average_Annual_Volume' ada di result_df sebelum melakukan nlargest
if 'Average_Annual_Volume' in result_df.columns:
    # Menyortir DataFrame berdasarkan rata-rata volume tertinggi dan ambil top 5
    top_5_stocks = result_df.nlargest(5, 'Average_Annual_Volume')
```
